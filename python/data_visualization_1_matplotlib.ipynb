### Line Plot

```python
## Importing the library
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
## Creating data for Line Plot
month = np.arange(1,13)
#price = sorted(list(np.random.randint(200,1000,12)))
price = sorted(list(np.random.randint(200,400,12)))
print(month)
print(price)
```

```python
##  Plotting the data
plt.plot(month, price, color='r',linestyle='--')

plt.title('Price trend of Products over the year')
plt.xlabel('Month', color='g')
plt.ylabel("Price", color='g')

plt.show()
```

```python
##  Plotting the data
#plt.plot(month, price, color='b', marker='*')
plt.plot(month, price,marker='.')
plt.title('Price trend over the year')
plt.xlabel('Month')
plt.ylabel("Price")

## To add a grid line to the plot
plt.grid(linestyle='--', linewidth='.5', color='r')
plt.show()
```

### Multiline plot

```python
## Reading the data for the plot
df_ins=pd.read_csv("Datasets/dataframe operations/insurance_data_with_day.csv")
print(df_ins.head())

#print(df_ins.dayofmonth.unique())

## Grouping the data by dayofmonth and getting the unique counts
df_ins_groupby = df_ins.groupby('dayofmonth')
days = df_ins_groupby['dayofmonth'].unique()
print(days)
```

```python
##  Printing average of each day of BMI, BP and Age in same plot

plt.plot(days, df_ins_groupby['bmi'].mean(), marker='.',color='green',label='BMI')
plt.plot(days, df_ins_groupby['bloodpressure'].mean(), marker='.',color='red',label='Blood Pressure')
plt.plot(days, df_ins_groupby['age'].mean(), marker='.',color='goldenrod',label='age')

plt.title('Multiline plots')
plt.xlabel('Day')
plt.ylabel('Measures')
plt.legend(loc='center left')
#plt.legend()
plt.show()
```

```python
## Generating the data
days = np.arange(1,31)
apr = ((np.random.random(30)) * 20).round(2)
may = ((np.random.random(30)) * 20).round(2)
jun = ((np.random.random(30)) * 20).round(2)

plt.plot(days,apr, label='April')
plt.plot(days,may, label='May')
plt.plot(days,jun, label='June')
plt.xlabel('Days')
plt.ylabel('Total Rainfall (cm)')
plt.title('Rainfalls in Apr, May and June')
#plt.legend()
plt.legend(bbox_to_anchor=[1,1])
plt.show()
```

```python
## Grouping the data using multi columns
df_ins_groupby_day_and_gender = df_ins.groupby(['dayofmonth','gender'])

print(df_ins_groupby_day_and_gender.mean()['claim'].head())

## Unstacking the gender
print(df_ins_groupby_day_and_gender.mean()['claim'].unstack().head())
```

```python
## Plotting the data
#plt.figuresize=(10,6)
df_ins_groupby_day_and_gender.mean()['claim'].unstack().plot(linewidth='2.0',marker='*')
plt.legend(loc='upper left')

plt.xlabel("Day")
plt.ylabel('Claim')
plt.title('Insurance claims by gender for every day of month',color='brown')
plt.show()
```

```python
df_ins_claim = df_ins_groupby_day_and_gender.mean()['claim'].unstack()
df_ins_claim.head()
```

```python
fig, ax = plt.subplots(1,2)

fig.set_figheight(5)  ## Setting the height of each plot
fig.set_figwidth(10)  ## Setting the width of each plot
fig.tight_layout(pad=1)  ## Increase the space between plots

ax[0].plot(df_ins_claim['male'])
ax[0].title.set_text('Male Claim')
ax[1].plot(df_ins_claim['female'])
ax[1].title.set_text('Female Claim')

plt.show()
```

### Subplots

```python
## Declaring the number of plots
fig, ax = plt.subplots(2,2)

x = np.linspace(0,8,1000)

fig.set_figheight(5)  ## Setting the height of each plot
fig.set_figwidth(10)  ## Setting the width of each plot
fig.tight_layout(pad=0.1)  ## Increase the space between plots

ax[0,0].plot(x, x ** 2,'g')
ax[0,0].title.set_text('Axis[0,0]')

ax[0,1].plot(x, x ** 3,'b')
ax[0,1].title.set_text('Axis[0,1]')

ax[1,0].plot(x,np.sin(x),'k')
ax[1,0].title.set_text('Axis[1,0]')

ax[1,1].plot(x,np.cos(x),'r')
ax[1,1].title.set_text('Axis[1,1]')

plt.show()
```

## Pie Plots

To display a portion of units and it is an univariate analysis

```python
df_ins.region.value_counts()
```

```python
## Creating data
freq_by_reg = df_ins.region.value_counts()
print(type(freq_by_reg))

## Keys
keys = list(freq_by_reg.keys())
print(keys)

## Counts
counts = freq_by_reg.to_list()
print(counts)
```

```python
## Plotting the data in pie chart
#plt.pie(counts, labels=keys)
plt.pie(counts, labels=keys, autopct='%1.1f%%')
plt.title('Regionwise Distribution')
plt.legend(bbox_to_anchor=[1,1])
plt.show()
```

```python
## Importing cm library
from matplotlib import cm

## Adding custom colors using cm module from matplotlib
cs = cm.Set1([1,3,5,7])
cs = cm.Set1([2,4,6,8])

plt.pie(counts, labels=keys, autopct='%1.1f%%', colors=cs)
plt.title('Regionwise Distribution')
plt.show()
```

```python
## Showing pie with highest value separately
explode=(.1,0,0,.1)

plt.pie(counts, labels=keys, explode=explode, autopct='%1.1f%%')
plt.title('Regionwise Distribution')
plt.show()
```

```python
## Adding circle to current pie chart

## Plotting the data in pie chart
plt.pie(counts, labels=keys, autopct='%1.1f%%')
plt.title('Regionwise Distribution')

circle = plt.Circle(xy=(0,0), radius=0.3, color='white')
plt.gca()

plt.gca().add_artist(circle)
plt.show()
```

### Scatter Plot

It is used to display relationship between 2 numerical variables.

```python
## plt.scatter() to display scatter plot
plt.scatter(x='bmi', y='claim', data=df_ins)

plt.title('Relationship between BMI and Claims')
plt.xlabel('BMI')
plt.ylabel('Claim')

plt.show()
```

```python
colors={"Yes":'red',"No":'green'}
## plt.scatter() to display scatter plot
plt.scatter(x='bmi', y='claim', data=df_ins, color=df_ins['smoker'].map(colors))

plt.title('Relationship between BMI and Claims')
plt.xlabel('BMI')
plt.ylabel('Claim')

plt.show()
```
